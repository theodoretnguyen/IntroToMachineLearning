```python
import numpy as np
```

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
A
```

```python
B = np.array([[7, 8], [9, 10], [11, 12]])
B
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(-1.0, 1.0, 0.005) * np.pi #numpy.arange(start, stop, step)
s = np.sin(x)
plt.title("sine function")
plt.grid(True)
plt.plot(x, s)
plt.show()
```
